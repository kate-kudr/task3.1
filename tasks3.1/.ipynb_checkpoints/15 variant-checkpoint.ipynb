## Текст задания

Реализовать разбор двоичного формата данных (в духе формата WAD игры Doom
или графического формата PNG). Данные начинаются с сигнатуры 0x5a 0x48 0x55 0x56, за которой
следует структура A. Порядок байт: от младшего к старшему. Адреса указаны в виде смещений
от начала данных. В решении разрешено использовать модуль struct.

## Реализация

```python
from struct import Struct, unpack_from, iter_unpack, calcsize


def f31(data):
    a = Struct('< I 2I Q H 4s B H I H')
    b = Struct('< i I H b')
    c = Struct('< b d')
    d = Struct('< d I 2b')

    u_a = a.unpack_from(data, data.index(b'ZHUV') + 4)
    u_b = b.unpack_from(data, u_a[0])
    u_c = [{'C1': c.unpack_from(data, u_a[i])[0],
            'C2': c.unpack_from(data, u_a[i])[1]} for i in range(1, 3)]
    u_d = d.unpack_from(data, u_a[-1])

    return {'A1': {'B1': u_b[0], 'B2': str(unpack_from(f'< {u_b[1]}s', data, u_b[2])[0])[2:-1], 'B3': u_b[3]},
            'A2': u_c,
            'A3': u_a[3],
            'A4': u_a[4],
            'A5': str(u_a[5])[2:-1],
            'A6': u_a[6],
            'A7': [x[0] for x in iter_unpack('< q', data[u_a[-2]:calcsize('< q') * u_a[-3] + u_a[-2]])],
            'A8': {'D1': u_d[0], 'D2': u_d[1], 'D3': u_d[2], 'D4': u_d[3]}
            }
```
