```python
#!/usr/bin/python3
from struct import *  # import struct

# """
d = (b'\xe7RVYN~=\x00\xf6\x8a\x0f\xff\x9e\xda\xafN9\x84\xb7\x01\xef\x1eI\xd4\xbfy)c'
     b'\x02\xfb\xde^\x9a\x94\x8d\xe4+z\x1f\xd8\x1f\xcd[\xd4\xf6\x14\x90G\xc14`\x02'
     b'^\xb1Cj\xdb#`l\xea\x97\x1e#\xd9\xe0\xaa\x9d\xa9k\xbbg7\xb1\xc2\x06'
     b'\x85\xbe\r\x9cD\xfe\xe1\x14\xc0(\x13(9\xdeDb{bXl.\xf9\xaf\xc9\x02,\x93\x02'
     b'\x00\x00\x00\t\x00\x00\x00\x96\xa9(\xcd\xf7\xa5\xae\xcb\x0bwl|>\x01\xdbN\xdb'
     b'\x9a+W\x19\xcf\x15\xad\xc1\xc8\xa6\x94\x00z&m\xed\x11\x00\xcc\x83'
     b':\x17\xfe\xc3\x8c>\x17o\xea3\xf3\x1b\x00')
"""
d=(b'\xe7RVYNL=\x00\xce\x8c\x1e#\x7fm&%a8 N\x97q&\xca?\xfb\x93\xfaD\xdb\xb4\x14'
b'iy\x91o$\x83\xc7\xcdZ\xd1\xe4\xc4\xfc\x82&Q\x17\xdc \xee$\xfb_\x1f\x0casM'
b'\xc5so\xac<\xb85\xa2io\xb1\xd8\xac\xb3\xe5g\xcb>\xa7?\xf2\x81G\xa99\xe99\t'
b'XNed3\xb4\xb8b\x81i\xa5C\x87\xe7\xbe\x02\x00\x00\x00\t\x00\x00\x00\x9c'
b'\xad\xdc\x1c\xaf\xef\x8c\xf3#gR\n?\xb5\x83p\x95\xbbHN7\xde\xa2\xd4\xd3'
b'\xd9\xf0\x13\xbdt^\xf1\xe6\x11\x00:\x82\xb8\xc6\xber\x83\xa4\xf5\xc3'
b'\xfb\x00\x94\x1b\x00')
#"""


def f31(I):
    if debug > 0: print("All Length=", len(I))
    # check signature
    # signature="{:x}".format(I[0])+"{:x}".format(I[1])+"{:x}".format(I[2])+"{:x}".format(I[3])+"{:x}".format(I[4])
    signature = ''.join(map("{:x}".format, I[0:5]))
    # 0xe7 0x52 0x56 0x59 0x4e,
    if signature == 'e75256594e':
        if debug > 0: print('OK_Signature', I[0], I[1], I[2], I[3], I[4], "=", signature)
    else:
        if debug > 0: print('ERROR_Signature', I[0], I[1], I[2], I[3], I[4], '<>', signature)
        return ()

    O = {}
    # A1 int8=b(1byte)
    offsetA = 5
    O["A1"] = unpack_from('b', I, offsetA)[0]  #
    if debug > 1: print("A1=", O["A1"], "offsetA=", offsetA)
    offsetA += 1

    # A2 Адрес (uint16=H(2byte)) структуры B
    O["A2"] = {}
    B_addr = unpack_from('<H', I, offsetA)[0]  # 0:4-signature=>offset=5 5:6 
    offsetA += 2
    offsetB = 0

    # B1 uint32=I(4byte)
    O["A2"]["B1"] = unpack_from('<I', I, B_addr)[0]
    if debug > 1: print("A2B1=", O["A2"]["B1"], "b=", I[B_addr:B_addr + 4], "B_addr=", B_addr, "offsetB=", offsetB)
    offsetB += 4
    # B2 Структура C
    O["A2"]["B2"] = {}

    # C1 uint8=B(1byte)
    O["A2"]["B2"]["C1"] = unpack_from('<B', I, B_addr + offsetB)[0]
    if debug > 1: print("A2B2C1=", O["A2"]["B2"]["C1"], "offsetB=", offsetB)
    offsetB += 1
    # C2 int64=q(8byte)
    O["A2"]["B2"]["C2"] = unpack_from('<q', I, B_addr + offsetB)[0]
    if debug > 1: print("A2B2C2=", O["A2"]["B2"]["C2"], "offsetB=", offsetB)
    offsetB += 8
    # C3 float=f(4byte)
    O["A2"]["B2"]["C3"] = unpack_from('<f', I, B_addr + offsetB)[0]
    if debug > 1: print("A2B2C3=", O["A2"]["B2"]["C3"], "offsetB=", offsetB)
    offsetB += 4
    # C4 Массив структур D, размер 2
    O["A2"]["B2"]["C4"] = []
    for i in range(2):  # 0..1
        O["A2"]["B2"]["C4"].append({})
        # D1 int64=q(8byte)
        O["A2"]["B2"]["C4"][i]["D1"] = unpack_from('<q', I, B_addr + offsetB)[0]
        if debug > 2: print("i=", i, "D1:=", O["A2"]["B2"]["C4"][i]["D1"], "offset=", offsetB)
        offsetB += 8
        # D2 int8=b(1byte)
        O["A2"]["B2"]["C4"][i]["D2"] = unpack_from('<b', I, B_addr + offsetB)[0]
        if debug > 2: print("i=", i, "D2:=", O["A2"]["B2"]["C4"][i]["D2"], "offset=", offsetB)
        offsetB += 1
        # O["A2"]["B2"]["C4"].append(D)
    if debug > 1: print("A2B2C4=", O["A2"]["B2"]["C4"], "offsetB=", offsetB)

    # C5 Структура E
    O["A2"]["B2"]["C5"] = {}
    # E1 int8=b(1byte)
    O["A2"]["B2"]["C5"]["E1"] = unpack_from('<b', I, B_addr + offsetB)[0]
    if debug > 1: print("A2B2C5E1=", O["A2"]["B2"]["C5"]["E1"], "offsetB=", offsetB)
    offsetB += 1
    # E2 int16=h(2byte)
    O["A2"]["B2"]["C5"]["E2"] = unpack_from('<h', I, B_addr + offsetB)[0]
    if debug > 1: print("A2B2C5E2=", O["A2"]["B2"]["C5"]["E2"], "offsetB=", offsetB)
    offsetB += 2
    # E3 int32=i(4byte)
    O["A2"]["B2"]["C5"]["E3"] = unpack_from('<i', I, B_addr + offsetB)[0]
    if debug > 1: print("A2B2C5E3=", O["A2"]["B2"]["C5"]["E3"], "offsetB=", offsetB)
    offsetB += 4
    # E4 Размер (uint32=I(4byte)) и адрес (uint32=I(4byte)) массива int32=i(4byte)
    O["A2"]["B2"]["C5"]["E4"] = []
    E4_sz = unpack_from('<I', I, B_addr + offsetB)[0]
    if debug > 2: print("E4_sz=", E4_sz, "offsetB=", offsetB)
    offsetB += 4

    E4_addr = unpack_from('<I', I, B_addr + offsetB)[0]
    if debug > 2: print("E4_addr=", E4_addr, "offsetB=", offsetB)
    offsetB += 4

    for i_E4 in range(E4_sz):  # 0..1
        O["A2"]["B2"]["C5"]["E4"].append(unpack_from('<i', I, E4_addr + i_E4 * 4)[0])
        if debug > 2: print("A2B2C5E4=", O["A2"]["B2"]["C5"]["E4"][i_E4])

    # E5 int64=q(8byte)
    O["A2"]["B2"]["C5"]["E5"] = unpack_from('<q', I, B_addr + offsetB)[0]
    if debug > 1: print("A2B2C5E5=", O["A2"]["B2"]["C5"]["E5"], "offsetB=", offsetB)
    offsetB += 8
    # E6 int8=b(1byte)
    O["A2"]["B2"]["C5"]["E6"] = unpack_from('<b', I, B_addr + offsetB)[0]
    if debug > 1: print("A2B2C5E6=", O["A2"]["B2"]["C5"]["E6"], "offsetB=", offsetB)
    offsetB += 1

    # C6 float=f(4byte)
    O["A2"]["B2"]["C6"] = unpack_from('<f', I, B_addr + offsetB)[0]
    if debug > 1: print("A2B2C6=", O["A2"]["B2"]["C6"], "offsetB=", offsetB)
    offsetB += 4
    # C7 Массив int32=i(4byte), размер 5
    O["A2"]["B2"]["C7"] = []
    for i in range(5):  # 0..4
        O["A2"]["B2"]["C7"].append(unpack_from('<i', I, B_addr + offsetB)[0])
        if debug > 2: print("i=", i, "A2B2C7=", O["A2"]["B2"]["C7"][i])
        offsetB += 4
    if debug > 1: print("A2B2C7=", O["A2"]["B2"]["C7"])
    # C8 Адрес (uint16=H(2byte)) структуры F
    O["A2"]["B2"]["C8"] = {}
    F_addr = unpack_from('<H', I, B_addr + offsetB)[0]
    if debug > 2: print("F_addr=", F_addr, "offsetB=", offsetB)
    offsetB += 2

    F = {}
    # F1 double=d(8byte)
    F["F1"] = unpack_from('<d', I, F_addr + 0)[0]
    if debug > 2: print("F1:=", F["F1"], "offsetF=0")
    # F2 int16=h(2byte)
    F["F2"] = unpack_from('<h', I, F_addr + 8)[0]
    if debug > 2: print("F2:=", F["F2"], "offsetF=8")
    O["A2"]["B2"]["C8"] = F
    if debug > 1: print("A2B2C8=", O["A2"]["B2"]["C8"], "offsetB=", offsetB)

    # B3 uint8=B(1byte)
    O["A2"]["B3"] = unpack_from('<B', I, B_addr + offsetB)[0]
    if debug > 1: print("A2B3=", O["A2"]["B3"], "b=", I[B_addr + offsetB:B_addr + offsetB + 1], "offsetB=", offsetB)
    offsetB += 1
    # B4 int32=i(4byte)
    O["A2"]["B4"] = unpack_from('<i', I, B_addr + offsetB)[0]
    if debug > 1: print("A2B4=", O["A2"]["B4"], "b=", I[B_addr + offsetB:B_addr + offsetB + 1], "offsetB=", offsetB)
    offsetB += 4
    # B5 int64=q(8byte)
    O["A2"]["B5"] = unpack_from('<q', I, B_addr + offsetB)[0]
    if debug > 1: print("A2B5=", O["A2"]["B5"], "b=", I[B_addr + offsetB:B_addr + offsetB + 1], "offsetB=", offsetB)
    offsetB += 8
    # B6 Адрес (uint16=H(2byte)) структуры G
    O["A2"]["B6"] = {}
    G_addr = unpack_from('<H', I, B_addr + offsetB)[0]
    if debug > 2: print("G_addr=", G_addr, "offsetB=", offsetB)
    offsetB += 2

    G = {}
    # G1 uint16=H(2byte)
    G["G1"] = unpack_from('<H', I, G_addr + 0)[0]
    if debug > 2: print("G1:=", G["G1"], "offsetG=0")
    # G2 uint32=I(4byte)
    G["G2"] = unpack_from('<I', I, G_addr + 2)[0]
    if debug > 2: print("G2:=", G["G2"], "offsetG=2")
    # G3 Массив int32=i(4byte), размер 7
    G["G3"] = []
    for i in range(7):  # 0..6
        G["G3"].append(unpack_from('<i', I, G_addr + 2 + 4 + i * 4)[0])
        if debug > 2: print("i=", i, "G3=", G["G3"][i])
    O["A2"]["B6"] = G
    if debug > 1: print("A2B6=", O["A2"]["B6"], "offsetB=", offsetB)

    # A3 int8=b(1byte)
    O["A3"] = unpack_from('b', I, offsetA)[0]
    offsetA += 1

    return (O)


debug = 0
if debug > 0: print("get:", d)
ret = f31(d)
print("f31:", str(ret))
```
