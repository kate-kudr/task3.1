```python
#!/usr/bin/python3
from struct import *  # import struct

# """
d = (b'NPQ\xa3\xc5\xc5;\x00F\x00Q\x00\\\x00g\x00r\x00}\x00\x02\x00\x00\x00'
     b'\x88\x00\x03\x00\x90\x00\x00\x00_\x19\xaey!-\xf5I\xd4y\xbc\x94o\xcb!\xeb'
     b'.#\xb0\xd1\x93\x00\x00\x00\x88y\xac\xec\x14\x146]\x9dp\xe23o\x1b\x1f\xae'
     b'\xc3l\xb9x/\xdfH\x0e/\xb8\n\x1b\xce,\xf3\x98\xb1\xe6\xd2Z>\x10 \xb1'
     b'6C\x0c\xdb\xdf\xd5\xcf5D\xda\xa9\xa9_\x0e\xd8\xe3\x96\xb5\xdd\xf3mVN#\xef.?$'
     b"\xc7'\xa0\x80\xdd\xed\xc7\xbah\x1b\xc3\r\xb0\xd5\x18\xbf~\xdeS>\x1fMu\x15"
     b'\x0b0\xbf\x19\xe1\xf6\xe5\x80\x1b\xabE\xbbb-\xba>r\xbb\xa8AO\x98')
"""
d=(b'NPQ\xa3\xaeI;\x00F\x00Q\x00\\\x00g\x00r\x00}\x00\x03\x00\x00\x00'
b'\x88\x00\x03\x00\x94\x00\x00\x00\x8d\x90\x91\x80e\x8f,b\xec\x92\x18\xcf'
b'C\x01\xcd\xf3W\xd5\x0c\xe1\x97\x00\x00\x00*#qQ\x08>5M\xfa<\x14\x10'
b'\xaf\xd58\xd5+\xc7\xdc\x80E\xfc\x9d\xaa\x93\x86\x97\xab?\x9b\x99\xeeP\x97./'
b'\xc9nnK\xb8\xed\xa7\xebyG\xb2%\x1b\x83\xb6\xc3\x93\xddT\x17\xb8\xb8\x92\xf2'
b'\xa10\xff\xa2\x94\x85\xc4\xdb\x87$|\xba\x8b\x8a\xfa"\x97U}Y\x1d\xc0o\xbf'
b'\r\xb1,<\x94`\x88=\xc8\xfc\xf8\xbc\xcf\xe1>\xb5T\x1dh\x85\xd12\xffY'
b'mL\x1c\xa2\xdc\xbd#\xd3%\xcc')
#"""


def f31(I):
    if debug > 0: print("All Length=", len(I))
    # check signature
    # signature="{:x}".format(I[0])+"{:x}".format(I[1])+"{:x}".format(I[2])
    signature = ''.join(map("{:02x}".format, I[0:4]))
    # 0x4e 0x50 0x51 0xa3
    if signature == '4e5051a3':
        if debug > 0: print('OK_Signature', I[0], I[1], I[2], I[3], "=", signature)
    else:
        if debug > 0: print('ERROR_Signature (', I[0], I[1], I[2], I[3], ")dec (", signature, ')hex <> 4e5051a3')
        return ()

    O = {}
    # A1 uint16=H(2bytes)
    offsetA = 4
    O["A1"] = unpack_from('<H', I, offsetA)[0]  #
    if debug > 1: print("A1=", O["A1"], "offsetA=", offsetA)
    offsetA += 2

    # A2 Массив адресов (uint16=H(2bytes)) структур B, размер 7
    O["A2"] = []
    for i in range(7):  # 0..6
        O["A2"].append({})

        B_addr = unpack_from('<H', I, offsetA)[0]
        if debug > 2: print("B_addr=", B_addr, "offsetA=", offsetA)
        offsetA += 2
        offsetB = 0

        # B1 uint64=Q(8bytes)
        O["A2"][i]["B1"] = unpack_from('<Q', I, B_addr + offsetB)[0]
        if debug > 2: print("B1=", O["A2"][i]["B1"], "offsetB=", B_addr + offsetB)
        offsetB += 8

        # B2 uint16=H(2bytes)
        O["A2"][i]["B2"] = unpack_from('<H', I, B_addr + offsetB)[0]
        if debug > 2: print("B2=", O["A2"][i]["B2"], "offsetB=", B_addr + offsetB)
        offsetB += 2

        # B3 int8=b(1byte)
        O["A2"][i]["B3"] = unpack_from('<b', I, B_addr + offsetB)[0]
        if debug > 2: print("B3=", O["A2"][i]["B3"], "offsetB=", B_addr + offsetB)
        offsetB += 1

    # A3 Структура C
    O["A3"] = {}

    # C1 Размер (uint32=I(4bytes)) и адрес (uint16=H(2bytes)) массива float=f(4bytes)
    O["A3"]["C1"] = []
    C1_sz = unpack_from('<I', I, offsetA)[0]
    if debug > 2: print("C1_sz=", C1_sz, "offsetA=", offsetA)
    offsetA += 4

    C1_addr = unpack_from('<H', I, offsetA)[0]
    if debug > 2: print("C1_addr=", C1_addr, "offsetA=", offsetA)
    offsetA += 2
    offsetC = 0

    for i_C1 in range(C1_sz):  # 0..x
        O["A3"]["C1"].append(unpack_from('<f', I, C1_addr + offsetC)[0])
        if debug > 2: print("A3C1=", O["A3"]["C1"][i_C1])
        offsetC += 4

    # C2 Размер (uint16=H(2bytes)) и адрес (uint32=I(4bytes)) массива int8=b(1byte)
    O["A3"]["C2"] = []
    C2_sz = unpack_from('<H', I, offsetA)[0]
    if debug > 2: print("C2_sz=", C2_sz, "offsetA=", offsetA)
    offsetA += 2

    C2_addr = unpack_from('<I', I, offsetA)[0]
    if debug > 2: print("C2_addr=", C2_addr, "offsetA=", offsetA)
    offsetA += 4
    offsetC = 0

    for i_C2 in range(C2_sz):  # 0..x
        O["A3"]["C2"].append(unpack_from('<b', I, C2_addr + offsetC)[0])
        if debug > 2: print("A3C2=", O["A3"]["C2"][i_C2])
        offsetC += 1

    # C3 uint64=Q(8bytes)
    O["A3"]["C3"] = unpack_from('<Q', I, offsetA)[0]
    if debug > 1: print("A3C3=", O["A3"]["C3"], "offsetA=", offsetA)
    offsetA += 8

    # C4 int64=q(8bytes)
    O["A3"]["C4"] = unpack_from('<q', I, offsetA)[0]
    if debug > 1: print("A3C4=", O["A3"]["C4"], "offsetA=", offsetA)
    offsetA += 8

    # C5 uint32=I(4bytes)
    O["A3"]["C5"] = unpack_from('<I', I, offsetA)[0]
    if debug > 1: print("A3C5=", O["A3"]["C5"], "offsetA=", offsetA)
    offsetA += 4

    # C6 Адрес (uint32=I(4bytes)) структуры D
    O["A3"]["C6"] = {}
    D_addr = unpack_from('<I', I, offsetA)[0]
    if debug > 2: print("D_addr=", D_addr, "offsetA=", offsetA)
    offsetA += 4
    offsetD = 0

    # D1 float=f(4bytes)
    O["A3"]["C6"]["D1"] = unpack_from('<f', I, D_addr + offsetD)[0]
    if debug > 1: print("A3C6D1=", O["A3"]["C6"]["D1"], "offsetD=", offsetD)
    offsetD += 4

    # D2 int8=b(1byte)
    O["A3"]["C6"]["D2"] = unpack_from('<b', I, D_addr + offsetD)[0]
    if debug > 1: print("A3C6D2=", O["A3"]["C6"]["D2"], "offsetD=", offsetD)
    offsetD += 1

    # D3 Массив uint16=H(2bytes), размер 4
    O["A3"]["C6"]["D3"] = []
    for i in range(4):  # 0..3
        O["A3"]["C6"]["D3"].append(unpack_from('<H', I, D_addr + offsetD)[0])
        if debug > 2: print("i=", i, "A3C6D3=", O["A3"]["C6"]["D3"][i])
        offsetD += 2
    if debug > 1: print("A3C6D3=", O["A3"]["C6"]["D3"])

    # D4 uint16=H(2bytes)
    O["A3"]["C6"]["D4"] = unpack_from('<H', I, D_addr + offsetD)[0]
    if debug > 1: print("A3C6D4=", O["A3"]["C6"]["D4"], "offsetD=", offsetD)
    offsetD += 2
    # D5 int64=q(8bytes)
    O["A3"]["C6"]["D5"] = unpack_from('<q', I, D_addr + offsetD)[0]
    if debug > 1: print("A3C6D5=", O["A3"]["C6"]["D5"], "offsetD=", offsetD)
    offsetD += 8

    # C7 int16=h(2bytes)
    O["A3"]["C7"] = unpack_from('<h', I, offsetA)[0]
    if debug > 1: print("A3C7=", O["A3"]["C7"], "offsetA=", offsetA)
    offsetA += 2

    # C8 uint8=B(1byte)
    O["A3"]["C8"] = unpack_from('<B', I, offsetA)[0]
    if debug > 1: print("A3C8=", O["A3"]["C8"], "offsetA=", offsetA)
    offsetA += 1

    return (O)


debug = 3
if debug > 0: print("get:", d)
ret = f31(d)
print("f31:", str(ret))
```
