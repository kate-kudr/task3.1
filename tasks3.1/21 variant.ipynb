```python
#!/usr/bin/python3
from struct import *  # import struct

# """
d = (b'ZPB\xba\xfb\x0e\x08\xdf\x9dE\xbaBD\x00\x8c\xa4\x99\xf5\xac\xcb'
     b'\x04\x07\xe4\x00\x00\x00\x04\x00\xa0\x86\xb3r\xf7\xf0f1\x00\xf35\x9d'
     b'\x10\xf1M\x8f#\x89\x0c[\xf82\xf8\xa6W\x1e\x00\x00\x00\x08\x00\xa4'
     b'\xff\xa0\x8eBG\x96\x1d\xd3\x90r\x00\xdd\xc3/\xd6\x95F\xf9\x1fX,\xcf\xe7\xc5'
     b"\xf7i\x12!\x15^/\xca\r}\x07\x12'\x0ci\xa9\xe46!\x81\xdc\x94\xe8\xb6"
     b'\x1c\xadD\xd0y\x03W\xc6G\x1e\xeaj{\x9a\xf4\xed\xba\xf9\x14\xbe\xfc6\xd87'
     b'4I\xe1\xc8\x97\x82\xd2\xb0?\xda\x19\xadN]U\x94\x00U\x00\x00\x00|o&'
     b">\x15\x15\xb4'\xff\xa9\xe7xgy$m\xc1\xb0\xc4")
"""
d=(b"ZPBI'\x9aD\xbbX\xa1\xe0\x84\xbe\x00\x8c\xd8]I\x86\x89N4b\x00\x00\x00\x03\x00"
b'\xa0\x045\xf3\xee\xcc\xf0g\xdd\xa2I\xf9!\x8a\x01\x81\xcbL\x1b\xda'
b'\xfd\x9e4\xfdE?\x00\x00\x00\x04\x00\xa3\xf14\x87^\x86\xf4\x9bwNv\x8f\x96'
b'\x16\x19\xcb#\xc1\xaf\xb6\x05\xb2L\xd6\xcd\xb0t\x9aW\x11\x83\xe8\x1e'
b'\x0f\xf8\xfatCI@v^\xffh\x07\x01\xe6\xba\x86\xe8}\xe8\x91y\r\xdc\xbd'
b'4\xc3m\x1b\x18\x98\x8eH\xec=j\xc97_\xf2P6"L\x89\xf9\xf1\xc8\x9d?\xda\xac\xb0'
b'\x91\xcb-\xe0\x00U\x00\x00\x00|\xd04\xbe\xdc\x99\x10\xaf\x8a\xfc;-\x1e\xf3')
#"""


def f31(I):
    if debug > 0: print("All Length=", len(I))
    # check signature
    # signature="{:x}".format(I[0])+"{:x}".format(I[1])+"{:x}".format(I[2])
    signature = ''.join(map("{:02x}".format, I[0:3]))
    # 0x5a 0x50 0x42,
    if signature == '5a5042':
        if debug > 0: print('OK_Signature', I[0], I[1], I[2], "=", signature)
    else:
        if debug > 0: print('ERROR_Signature (', I[0], I[1], I[2], ")dec (", signature, ')hex <> 5a5042')
        return ()

    O = {}
    # A1 uint64=Q(8bytes)
    offsetA = 3
    O["A1"] = unpack_from('>Q', I, offsetA)[0]  #
    if debug > 1: print("A1=", O["A1"], "offsetA=", offsetA)
    offsetA += 8

    # A2 int16=h(2bytes)
    O["A2"] = unpack_from('>h', I, offsetA)[0]  #
    if debug > 1: print("A2=", O["A2"], "offsetA=", offsetA)
    offsetA += 2

    # A3 Адрес (uint16=H(2bytes)) структуры B
    O["A3"] = {}
    B_addr = unpack_from('>H', I, offsetA)[0]  # 0:4-signature=>offset=5 6:7 
    offsetA += 2
    offsetB = 0

    # B1 double=d(8bytes)
    O["A3"]["B1"] = unpack_from('>d', I, B_addr + 0)[0]
    if debug > 2: print("B1:=", O["A3"]["B1"], "offsetB=0")
    offsetB += 8

    # B2 Адрес (uint16=H(2bytes)) структуры C
    O["A3"]["B2"] = {}
    C_addr = unpack_from('>H', I, B_addr + offsetB)[0]
    if debug > 2: print("C_addr=", C_addr, "offsetB=", offsetB)
    offsetB += 2
    offsetC = 0

    # C1 uint64=Q(8bytes)
    O["A3"]["B2"]["C1"] = unpack_from('>Q', I, C_addr + offsetC)[0]
    if debug > 1: print("A3B2C1=", O["A3"]["B2"]["C1"], "offsetC=", offsetC)
    offsetC += 8

    # C2 Структура D
    O["A3"]["B2"]["C2"] = {}

    # D1 int8=b(1byte)
    O["A3"]["B2"]["C2"]["D1"] = unpack_from('>b', I, C_addr + offsetC)[0]
    if debug > 1: print("A3B2C2D1=", O["A3"]["B2"]["C2"]["D1"], "offsetC=", offsetC)
    offsetC += 1

    # D2 uint16=H(2bytes)
    O["A3"]["B2"]["C2"]["D2"] = unpack_from('>H', I, C_addr + offsetC)[0]
    if debug > 1: print("A3B2C2D2=", O["A3"]["B2"]["C2"]["D2"], "offsetC=", offsetC)
    offsetC += 2

    # D3 int64=q(8bytes)
    O["A3"]["B2"]["C2"]["D3"] = unpack_from('>q', I, C_addr + offsetC)[0]
    if debug > 1: print("A3B2C2D3=", O["A3"]["B2"]["C2"]["D3"], "offsetC=", offsetC)
    offsetC += 8

    # D4 Массив uint16=H(2bytes), размер 6
    O["A3"]["B2"]["C2"]["D4"] = []
    for i in range(6):  # 0..5
        O["A3"]["B2"]["C2"]["D4"].append(unpack_from('>H', I, C_addr + offsetC)[0])
        if debug > 2: print("i=", i, "D4=", O["A3"]["B2"]["C2"]["D4"][i])
        offsetC += 2
    if debug > 1: print("A3B2C2D4=", O["A3"]["B2"]["C2"]["D4"], "offsetC=", offsetC)

    # D5 int64=q(8bytes)
    O["A3"]["B2"]["C2"]["D5"] = unpack_from('>q', I, C_addr + offsetC)[0]
    if debug > 1: print("A3B2C2D5=", O["A3"]["B2"]["C2"]["D5"], "offsetC=", offsetC)
    offsetC += 8

    # B3 Структура E
    O["A3"]["B3"] = {}

    # E1 Адрес (uint32=I(4bytes)) структуры F
    O["A3"]["B3"]["E1"] = {}
    F_addr = unpack_from('>I', I, B_addr + offsetB)[0]  # 0:4-signature=>offset=5 6:7
    if debug > 2: print("F_addr=", F_addr, "offsetB=", offsetB)
    offsetB += 4
    offsetF = 0

    F = {}
    # F1 Массив int16=h(2bytes), размер 3
    O["A3"]["B3"]["E1"]["F1"] = []
    for i in range(3):  # 0..2
        O["A3"]["B3"]["E1"]["F1"].append(unpack_from('>h', I, F_addr + offsetF)[0])
        if debug > 2: print("i=", i, "A3B3E1F1=", O["A3"]["B3"]["E1"]["F1"][i])
        offsetF += 2
    if debug > 1: print("A3B3E1F1=", O["A3"]["B3"]["E1"]["F1"])

    # F2 uint16=H(2bytes)
    O["A3"]["B3"]["E1"]["F2"] = unpack_from('>H', I, F_addr + offsetF)[0]
    if debug > 2: print("F2:=", O["A3"]["B3"]["E1"]["F2"], "offsetF=", offsetF)
    offsetF += 2

    # F3 int32=i(4bytes)
    O["A3"]["B3"]["E1"]["F3"] = unpack_from('>i', I, F_addr + offsetF)[0]
    if debug > 2: print("F3:=", O["A3"]["B3"]["E1"]["F3"], "offsetF=", offsetF)
    offsetF += 4

    # F4 uint32=I(4bytes)
    O["A3"]["B3"]["E1"]["F4"] = unpack_from('>I', I, F_addr + offsetF)[0]
    if debug > 2: print("F4:=", O["A3"]["B3"]["E1"]["F4"], "offsetF=", offsetF)
    offsetF += 4

    # E2 int16=h(2bytes)
    O["A3"]["B3"]["E2"] = unpack_from('>h', I, B_addr + offsetB)[0]
    if debug > 1: print("A3B3E2=", O["A3"]["B3"]["E2"], "offsetB=", offsetB)
    offsetB += 2

    # B4 float=f(4bytes)
    O["A3"]["B4"] = unpack_from('>f', I, B_addr + offsetB)[0]
    if debug > 1: print("A3B4=", O["A3"]["B4"], "offsetB=", offsetB)
    offsetB += 4

    # A4 int64=q(8bytes)
    O["A4"] = unpack_from('>q', I, offsetA)[0]
    if debug > 1: print("A4=", O["A4"], "offsetA=", offsetA)
    offsetA += 8

    # A5 Массив структур G, размер 2
    O["A5"] = []
    for i in range(2):  # 0..1
        O["A5"].append({})
        # G1 Размер (uint32=I(4bytes)) и адрес (uint16=H(2bytes)) массива int8=b(1byte)
        O["A5"][i]["G1"] = []
        G1_sz = unpack_from('>I', I, offsetA)[0]
        if debug > 2: print("G1_sz=", G1_sz, "offsetA=", offsetA)
        offsetA += 4

        G1_addr = unpack_from('>H', I, offsetA)[0]
        if debug > 2: print("G1_addr=", G1_addr, "offsetA=", offsetA)
        offsetA += 2

        for i_G1 in range(G1_sz):  # 0..x
            O["A5"][i]["G1"].append(unpack_from('>b', I, G1_addr + i_G1 * 1)[0])
            if debug > 2: print("A5iG1=", O["A5"][i]["G1"][i_G1])

        # G2 Массив (uint64=Q(8bytes)), размер 3
        O["A5"][i]["G2"] = []
        for j in range(3):  # 0..2
            O["A5"][i]["G2"].append(unpack_from('>Q', I, offsetA)[0])
            if debug > 2: print("i=", i, "A5iG2=", O["A5"][i]["G2"][j])
            offsetA += 8
        if debug > 1: print("A5iG2=", O["A5"][i]["G2"])

        # G3 uint8=B(1byte)
        O["A5"][i]["G3"] = unpack_from('>B', I, offsetA)[0]
        if debug > 1: print("A5iG3=", O["A5"][i]["G3"], "offsetA=", offsetA)
        offsetA += 1

    return (O)


debug = 3
if debug > 0: print("get:", d)
ret = f31(d)
print("f31:", str(ret))
```
